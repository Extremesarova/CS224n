**Question**: Briefly explain in at most three sentences what you see in the plot.  
**Answer**: I can see several clusters on the plot. For example, the words *tea* and *coffee* are located close to each other, which is expected. Words *female*, *woman* and *man* are also close to each other, although *male* is located apart from them. Also, I think the relationship king:male::queen:female should be true for this vectors (maybe word *female* will be on the second place after the word *man*). It is worth to notice that sometimes antonyms are very close to each other, like *enjoyable*, *annoying* and *sweet*; *great*, *boring*, *wonderful* and *amazing*. I think it happens due to the nature of the dataset (sentiment treebank).
